```python
# Counting Sort
def counting_sort_asec(A, B, k):
    C = [0] * (k+1)
    
    for i in A:
        C[i] += 1
    
    for i in range(1, len(C)):
        C[i] += C[i-1]

    for i in range(len(A)-1, -1, -1):
        C[A[i]] -= 1
        B[C[A[i]]] = A[i]
    
    return B
print(counting_sort_asec([0, 4, 1, 3, 1, 2, 4, 1], [0]*8, 8))

def counting_sort_desc(A, B, k):
    C = [0] * (k+1)
    
    for i in A:
        C[i] += 1
    
    # 8, 7, 4, 3, 2
    for i in range(len(C)-2, -1, -1):
        C[i] += C[i+1]
    
    for i in range(len(A)-1, -1, -1):
        C[A[i]] -= 1
        B[C[A[i]]] = A[i]
    return B

print(counting_sort_desc([0, 4, 1, 3, 1, 2, 4, 1], [0]*8, 8))
```

```python
# 🥈1018번 체스판 다시 칠하기
def chess_board(arr, start, other):
    re_color = 0
    for i in range(0, 8, 2):
        # 짝수 줄
        if arr[i] != start:
            re_color += 1
        # 홀수 줄
        if arr[i+1] != other:
            re_color += 1
    return re_color
        
n, m = map(int,input().split())
chess = []

for i in range(n):
    chess.append(list(input()))

n_diff, m_diff = n - 8, m-8
cnt = []

for i in range(n_diff+1):
    for j in range(m_diff+1):
        # 첫째줄 시작이 w일때와 b일때를 구해 비교
        start_w, start_b = 0, 0
        # 짝수 줄
        for k in range(i, 8+i, 2):
            # 첫째줄 시작이 w이면 짝수번째 줄은 모두 w로 시작
            start_w += chess_board(chess[k][j:j+8], "W", "B")
            start_b += chess_board(chess[k][j:j+8], "B", "W")
        # 홀수 줄
        for k in range(i+1, 8+i, 2):
            # 첫째줄 시작이 w이면 짝수번째 줄은 모두 b로 시작
            start_w += chess_board(chess[k][j:j+8], "B", "W")
            start_b += chess_board(chess[k][j:j+8], "W", "B")
        # 두 값중에 더 작은 값 반환
        cnt.append(start_b if start_b < start_w else start_w)

print(min(cnt))
```

```python
def binary_search(arr, n, key):
    # 검색범위 지정
    # 시작, 끝
    start = 0
    end = n-1

    while True:
        mid = (start + end)//2
        if arr[mid] == key:
            print("find")
            return mid
        elif arr[mid] > key:
            end = mid -1
        elif arr[mid] < key:
            start = mid + 1
        
        print("can't find")
        return -1
    
    

arr = [2,3,4,5,6,7,8,10]
n = len(arr)

print(binary_search(arr, n, 1))
```

```python
def selection_sort(arr, n):
    # 맨 앞자리 부터 자리의 주인을 정해준다. 작은 수자부터 차례대로
    for i in range(n-1):
        min_idx = i
        # i의 뒤부터 비교를 시작하면서 최솟값을 찾는다.
        for j in range(i+1, n):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
    return arr

arr = [2,3,4,1, 0, 5,12, 6,7,8,10]
n = len(arr)

print(selection_sort(arr, n))
```

```python
# 보이어무어

def bm_Search(txt, pat):
    n = len(txt)
    m = len(pat)
    
    # txt인덱스
    ti = m-1
    # pat인덱스
    pi = 0

    skip = {}

    for p in pat:
        # 같은 글자가 여러번이면 가장 오른쪽에 맞춰서
        skip[p] = m - 1 - pat.rfind(p)
    
    while ti < n:
        pi = m-1
        # 같은 문자가 나오면 계속 앞으로 이동
        while txt[ti] == pat[pi]:
            if pi == 0:
                # 패턴 모두 일치
                # 기준 위치인 ti를 리턴
                return ti
            pi -= 1
            ti -= 1
        
        # 다른 문자가 나오면 건너뛰기 표를 참고해서 skip
        # 딕셔너리에 없는 경우, 패턴에 없다는 뜻 ==> 패턴의 길이만큼 건너뛴다.
        offset = skip.get(txt[ti]) if skip.get(txt[ti]) else m
        ti += offset if offset > m-pi else m
    
    return -1

t = 'zzzabcdabcdabcefabcd'
p = 'abcdabcef'

print(f"res: ", bm_Search(t, p), t.find(p))
```

```python
# swea 물놀이를 가자
from collections import deque

dx = [-1, 1, 0, 0]
dy = [0,0,-1,1]

t = int(input())

for tc in range(1, t+1):
    n, m = map(int, input().split())
    result = [[-1] * m for _ in range(n)]

    q = deque()
    for i in range(n):
        row = list(input())
        for j in range(m):
            if row[j] == 'W':
                q.append((i, j))
                result[i][j] = 0
    
    sum_d = 0
    while q:
        ci, cj = q.popleft()
        for d in range(4):
            ni, nj = ci + dx[d], cj + dy[d]
            if 0 <= ni < n and 0 <= nj < m and result[ni][nj] == -1:
                result[ni][nj] = result[ci][cj] + 1
                q.append((ni, nj))
                sum_d += result[ni][nj]
    
    print(f"#{tc} {sum_d}")
```

```python
# 🥈 1244번 스위치 켜고 끄기
import sys
n = int(sys.stdin.readline().strip())
switch = ["0"] + list(sys.stdin.readline().strip().split())

m = int(sys.stdin.readline().strip())

# 남학생 : 1, 여학생 : 2
# 남학생 : 자기가 받은 수의 배수이면 바꾼다
# 여학생 : 자기가 받은 수를 중심으로 좌우가 대칭이면서 가장 많은 스위치를 포함하는 구간

for i in range(m):
    s, l = map(int, sys.stdin.readline().strip().split())
    if s == 1:
        for j in range(1, n+1):
            if j % l == 0:
                switch[j] = "0" if switch[j] == "1" else "1"
    else:
        switch[l] = "0" if switch[l] == "1" else "1"
        for j in range(1, l):
            f, t = l-j, l+j
            if 0 < f <= n and 0 < t <= n and switch[f] == switch[t]:
                switch[f] = "0" if switch[f] == "1" else "1"
                switch[t] = "0" if switch[t] == "1" else "1"
            else:
                break

for i in range(n//20+1):
    if n < 20*(i+1)+1:
        print(" ".join(switch[20*i+1:]))
    else:
        print(" ".join(switch[20*i+1:20*(i+1)+1]))
```
