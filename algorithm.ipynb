```python
# Counting Sort
def counting_sort_asec(A, B, k):
    C = [0] * (k+1)
    
    for i in A:
        C[i] += 1
    
    for i in range(1, len(C)):
        C[i] += C[i-1]

    for i in range(len(A)-1, -1, -1):
        C[A[i]] -= 1
        B[C[A[i]]] = A[i]
    
    return B
print(counting_sort_asec([0, 4, 1, 3, 1, 2, 4, 1], [0]*8, 8))

def counting_sort_desc(A, B, k):
    C = [0] * (k+1)
    
    for i in A:
        C[i] += 1
    
    # 8, 7, 4, 3, 2
    for i in range(len(C)-2, -1, -1):
        C[i] += C[i+1]
    
    for i in range(len(A)-1, -1, -1):
        C[A[i]] -= 1
        B[C[A[i]]] = A[i]
    return B

print(counting_sort_desc([0, 4, 1, 3, 1, 2, 4, 1], [0]*8, 8))
```

```python
def binary_search(arr, n, key):
    # ê²€ìƒ‰ë²”ìœ„ ì§€ì •
    # ì‹œì‘, ë
    start = 0
    end = n-1

    while True:
        mid = (start + end)//2
        if arr[mid] == key:
            print("find")
            return mid
        elif arr[mid] > key:
            end = mid -1
        elif arr[mid] < key:
            start = mid + 1
        
        print("can't find")
        return -1
    
    

arr = [2,3,4,5,6,7,8,10]
n = len(arr)

print(binary_search(arr, n, 1))
```

```python
def selection_sort(arr, n):
    # ë§¨ ì•ìë¦¬ ë¶€í„° ìë¦¬ì˜ ì£¼ì¸ì„ ì •í•´ì¤€ë‹¤. ì‘ì€ ìˆ˜ìë¶€í„° ì°¨ë¡€ëŒ€ë¡œ
    for i in range(n-1):
        min_idx = i
        # iì˜ ë’¤ë¶€í„° ë¹„êµë¥¼ ì‹œì‘í•˜ë©´ì„œ ìµœì†Ÿê°’ì„ ì°¾ëŠ”ë‹¤.
        for j in range(i+1, n):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
    return arr

arr = [2,3,4,1, 0, 5,12, 6,7,8,10]
n = len(arr)

print(selection_sort(arr, n))
```

```python
# swea ê³„ì‚°ê¸° 2
isp = {'+':1,'-':1,'*':2,'/':2} 

def get_post(string):
    post = ""
    stack = []
    for s in string:
        if "0" <= s <= "9":
            post += s
        else:
            while stack and isp[stack[-1]] >= isp[s]:
                post += stack.pop()
            stack.append(s)
            
    while stack:
        post += stack.pop()
    
    return post

def get_result(post):
    stack = []
    for p in post:
        if "0" <= p <= "9":
            stack.append(int(p))
        else:
            if len(stack) >= 2:
                second = stack.pop()
                first = stack.pop()
                if p == "+":
                    result = first + second
                elif p == "-":
                    result = first - second
                elif p == "*":
                    result = first * second
                else:
                    result = first // second
                stack.append(result)
    return stack.pop()

for tc in range(1, 11):
    n = int(input())
    string = input()
    print(f"#{tc} {get_result(get_post(string))}")
```
